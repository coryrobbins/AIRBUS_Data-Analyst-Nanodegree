## Boxplots

We're going to create boxplots for each country, but only for the top 10 countries in terms of some statistic

For now we're going to use the mean value for each country for the soring 

```python
top_10_growers = (co2.max() - co2.min()) / co2.min()
top_10_growers
```

```python
top_10_columns = {}
for name, df in all_data.items():
    top_10 = df.mean().dropna().sort_values().iloc[-10:]
    top_10_columns[name] = top_10.index
    
    
```

```python
top_10_columns
```

```python
# Digression: enumerate
print(list(enumerate(['a', 'b', 'c'], start=1)))
```

```python
plt.figure(figsize=(5, 30))
for num, name_df in enumerate(all_data.items(), start=1):
    name, df = name_df
    plt.subplot(10,2,num)
    df[top_10_columns[name]].boxplot()
    plt.title(name)
```

# Tranforming the data 

Now we want a flatten (tall and skinny) representation of the data 
to be more similar to other online resoources.

For this we need the stack / unstack operations. 

```python
all_data.keys()
all_data['continent'] = continents #['Continent']
# means['continent'] = continents['Continent']
```

```python
# all_data["pop"].transpose().head()
#all_data["continent"]
#all_data["continent"].head()
#all_data["pop"].head()
```

```python
combined = pd.DataFrame()
for name in all_data:
    datanow = (
        all_data[name]
            .stack()
            .reset_index()
            .rename(columns={"level_0":"year", 0:name})
            .set_index(["year", "country"]) #.transpose()
        )
    combined[name] = datanow[name] 

combined = combined.reset_index()
#combined = pd.DataFrame(combined)
```
